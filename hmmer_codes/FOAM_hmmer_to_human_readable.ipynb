This is the main code. As hmmer changes it's outfile you may need to modify this in the future.
You may want to change the minscore as best practices evolve. Depending on the genes we drop the score as low as 20. 

```python
column = 14 #default is 14, i.e. the p/dom score
minscore = 25 #"the min score required to keep the hit, default is 25"

HMMRres = "hmmer_hits.sort" #Your output file from hmmsearch
print("HMMERres: " + HMMRres)
col = column - 1
print("cols: " + str(col))
minscore = float(minscore)
print("minscore: " + str(minscore))

try:
	Hits=open(HMMRres, 'r')
except IOError as e:
	print("HMM output file not found or unreadable:"),HMMRres
	pass

best_hits = open("best_hits.txt","w")

while True:
    

	bestHit=Hits.readline()
	#print "bestHit" + bestHit
	if not bestHit.startswith("#"):	
		maxscore=float(bestHit.split()[col])
		#print "maxscore: " + str(maxscore)
		best_hits.write("maxscore: " + str(maxscore) + "\n")
		query=bestHit.split()[0]
		break

for hit in Hits:
	#print hit
	if not hit.startswith("#"):
		hitSplit=hit.split()
	else: continue
	curScore=float(hitSplit[col])
	#print "curScore: " + str(curScore)
	best_hits.write("curScore: " + str(curScore) + "\n")
	if curScore >= minscore:
		if query!=hitSplit[0]:
			if float(bestHit.split()[col]) >= minscore :
				#print bestHit,
				best_hits.write(bestHit)
			query=hitSplit[0]
			maxscore=float(0)
			bestHit=hit
		elif maxscore<=float(hitSplit[col]):
			bestHit=hit
			maxscore=float(hitSplit[col])
if bestHit is not None:
	#print "\n" + bestHit,
    print("Done processing best hits")

Hits.close()
best_hits.close()
```

```python
from collections import Counter

tids = "hmmsearch.BH.tmp1"
ac=open(tids, 'r')

gene_name = open("hmmsearch.BH.tmp2",'w')

	
lignes = ac.readlines()
ac.close()

mytids=[]

for i in lignes:
	mytids.append(str.strip(i,"\n"))


cnt = Counter()
for word in mytids:
	cnt[word] += 1

for i in cnt.keys():
	print(str(i)+"\t"+str(cnt[i]))
	gene_name.write(str(i)+"\t"+str(cnt[i])+"\n")
    
gene_name.close()
```
